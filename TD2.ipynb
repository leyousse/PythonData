```python
nat = pd.read_csv("Datasets/nat2018.csv", sep=";")
nat
```

```python
nat=nat[nat.annais!="XXXX"]
```

```python
nat
```

```python
#Proportion des naissances avec un prénom par rapport à la somme des naissances pour cette année et ce sexe

def add_prop(group):
    group['prop'] = group.nombre / group.nombre.sum()
    return group

nat = nat.groupby(['annais','sexe']).apply(add_prop)

nat.sort_values(by="prop")
```

```python
#SANITY CHECK 1:
nat.groupby(['annais','sexe']).prop.sum()
```

```python
#SANITY CHECK 2:
sums1900=nat[nat.annais=="1900"].groupby('sexe').nombre.sum()
sums1900
```

```python
print(48713/237650)
```

```python
#SANITY CHECK 2' (automatic, factorized):
# année du prénom avec la plus grande proportion
YearMaxProp = nat.sort_values(by="prop")[-1:].annais
YearMaxProp
```

```python
#ILOC permet d'acceder au contenu d'une celulle en transmettant ses coordonnées
YearMaxProp = nat.sort_values(by="prop")[-1:].annais.iloc[0]
#YearMaxProp=int(YearMaxProp)
YearMaxProp
```

```python
#nombre de naissances du genre et de l'année du prénom qui a eu la plus forte proportion

birthMaxProp = nat.sort_values(by="prop")[-1:].nombre.iloc[0]
birthMaxProp
```

```python
#max proportion
maxprop=nat.sort_values(by="prop")[-1:].prop.iloc[0]
maxprop
```

```python
#somme totale des naissances en 1900 dans le genre de Marie
nat[nat.annais==YearMaxProp].groupby(["annais","sexe"]).nombre.sum()
```

```python
sumsYearMaxProp=nat[nat.annais==YearMaxProp].groupby(["annais","sexe"]).nombre.sum().iloc[1]
sumsYearMaxProp
```

```python
if birthMaxProp / sumsYearMaxProp == maxprop :
    print("Sanity check 2 ... PASSED")
```
